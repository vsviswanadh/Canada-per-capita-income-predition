```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head()
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='*')
plt.xlabel('year')
plt.ylabel('per capita income in us dollars')
plt.title('Per Capita Income in Canada')
plt.show()
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
m=reg.coef_
```

```python
b=reg.intercept_
```

```python
y=m*2020+b
y
```

```python
df1=pd.read_csv('year.csv')
```

```python
df1
```

```python
predicted_income=reg.predict(df1)
```

```python
predicted_income
```

```python
df1['predicted per capita income (US$)'] = predicted_income
```

```python
df1.to_csv('predicted_per_capita_income.csv', index=False)
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='*')
plt.plot(df['year'],reg.predict(df[['year']]),color='blue', linewidth=2)
plt.xlabel('year')
plt.ylabel('per capita income in us dollars')
plt.title('Per Capita Income in Canada')
plt.show()
```
